# Initialization

Imports and initialization, read and track files:
Returned output displays the initial data provided in train.csv

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier


from sklearn import preprocessing
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
p_le = preprocessing.LabelEncoder()

train.head(20)
####Original data displayed in a table
```

# Data Cleaning

Exclude columns which will skew data ( ticket, name and cabin).
For numerical columns, if missing field replace it with the mean of its own column.
For columns containing Char (Embarked), if missing field replace with placeholder 'X'.
Returned output displays first 10 rows of the data that is now trimmed and much easier to work with:

```python
######################
def clean(train): #clean the data to make it easier to use/read
    train = train.drop(["Ticket", "Name", "Cabin"], axis = 1)#get rid of useless columns, wont effect result
    
    #ensure numerical fields are all complete
    #if numerical field is left blank, replace with mean of column
    train.Age.fillna(train.Age.mean(), inplace = True)
    train.SibSp.fillna(train.SibSp.mean(), inplace = True)
    train.Parch.fillna(train.Parch.mean(), inplace = True)
    train.Fare.fillna(train.Fare.mean(), inplace = True)
    train.Embarked.fillna("X", inplace = True)#if Embarked status left blank replace with "X" to represent no input
    train.Sex = p_le.fit_transform(train.Sex)#only actually want to transform data in test. train is used just to learn and reference
    test.Sex = p_le.transform(test.Sex)
    train.Embarked = p_le.fit_transform(train.Embarked)
    test.Embarked = p_le.transform(test.Embarked)
   
    return train

train = clean(train)
test = clean(test)
train.head(10)
######Cleaned data in table. Easy to see much more ledgable than before
```

# Decision Tree Algorithm

Drop Survived from training column and store.
Divide data into training and test sets:


```python
#divide data into training and testing sets
training_table = train.drop("Survived", axis = 1)
target_column = train["Survived"]

####from sklearn.model_selection import train_test_split
training_table_train, training_table_test, target_column_train, target_column_test = train_test_split(training_table, target_column, test_size=0.20)
```

Algorithm training, prediction adn accuracy test
Returned output is accuracy of predictions performed against true values for these fields approx: 77% accuracy:

```python
# train the algorithm

classifier = DecisionTreeClassifier()
classifier.fit(training_table_train, target_column_train)

#look at training data and make prediction for "Survived"
column_prediction = classifier.predict(training_table_test)

#eval and test accuracy
print(confusion_matrix(target_column_test, column_prediction))
print(classification_report(target_column_test, column_prediction))
```

# K- Nearest Neighbors Algorithm

Drop Survived from training column and store.
Divide data into training and test sets:


```python
###############KN

X = train.drop("Survived", axis = 1)
y = train["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

Feature scaling:

```python
####from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

Algorithm training, prediction adn accuracy test
Returned output is accuracy of predictions performed against true values for these fields approx: 78% accuracy:

```python
#Algorithm training
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)

#look at training data and make prediction for "Survived"
y_pred = clf.predict(X_test)

#eval and test accuracy
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Predictions

Apply each algorithm to test.csv and return PassengerID with predicted Survived status.
Returnded output is 2 files, one representing each of the 2 algorithm's' predictions on the status of each passenger.

```python
DTC_pred = classifier.predict(test)
return_file = pd.DataFrame({"PassengerId": test['PassengerId'], "Survived": DTC_pred})
return_file.to_csv("Titanic_DTC_prediction.csv")
```

```python
KNN_pred = clf.predict(test)
return_file = pd.DataFrame({"PassengerId": test['PassengerId'], "Survived": KNN_pred})
return_file.to_csv("Titanic_KNN_prediction.csv")
```
